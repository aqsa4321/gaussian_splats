```python
import open3d as o3d

def read_and_show_ply(file_path):
    # Load the point cloud from the PLY file
    point_cloud = o3d.io.read_point_cloud(file_path)
    
    # Print point cloud details (e.g., number of points, xyz coordinates, and optional RGB colors)
    print("Point Cloud Information:")
    print(point_cloud)
    
    # Check if the point cloud has colors
    if point_cloud.has_colors():
        print("The point cloud has RGB values for each point.")
    
    # Check if the point cloud has normals
    if point_cloud.has_normals():
        print("The point cloud has normal vectors for each point.")
    
    # Visualize the point cloud
    o3d.visualization.draw_geometries([point_cloud])

# Example usage:
file_path = '/home/csslab/Documents/Aqsa/jupyter/points3D_colmap.ply'  # Replace with the actual file path
read_and_show_ply(file_path)
```

```python
import open3d as o3d
import numpy as np

# Load the PLY file
ply_file = "/home/csslab/Documents/Aqsa/jupyter/points_mast3r.ply"
point_cloud = o3d.io.read_point_cloud(ply_file)

# Get the color data (red, green, blue)
colors = np.asarray(point_cloud.colors)  # Colors will be in range [0, 1]

# Convert color values to range [0, 255] if necessary
colors = (colors * 255).astype(int)

# Count how many points are black (R=0, G=0, B=0)
black_points_count = np.sum(np.all(colors == [0, 0, 0], axis=1))

print(f"Number of black points: {black_points_count}")
```
